### Airports

#### 2) How many airports are there altogether?

```python
pd.read_sql("""
    SELECT
        count(*)
    FROM
        airports
    """, con=db)
```

#### 3) How many airports are there in Morocco?

```python
pd.read_sql("""
    SELECT
        count(name)
    FROM
        airports
    where city = 'Morocco'
    """, con=db)
```

#### 5) Which country has the most airports?

```python
pd.read_sql("""
    SELECT 
        count(name), country
    FROM
        airports
    Group by country
    order by 1 desc
    limit 1
    """, con=db)
```

#### 6) What are the highest and lowest altitudes at which there are airports?

```python
pd.read_sql("""
 Select * from (
     SELECT
        'max altitude' as type, city, altitude
    FROM
        airports
    ORDER BY 3 DESC
    limit 1
    )
    
    union all
 
 Select * from (
 
    SELECT 'min altitude' as type, city, altitude
    from airports
    order by 3 asc
    limit 1
 )   
    """, con=db)
```

## Airlines

#### 7) Select the last 5 airlines alphabetically

```python
pd.read_sql("""
select * from 
(
select name from airlines order by trim(name) desc 
limit 5
) 
order by 1


""", con=db)
```

#### 8) Which country has the most *inactive* airlines?

```python
pd.read_sql("""
    select count(name) qty, country    
    from airlines
    where active = 'N'
    group by country 
    order by 1 desc 
    limit 1
    """, con=db)
```

#### 9) Which countries have *exactly* 9 active airlines?

Hint: look at the [HAVING clause](https://www.w3schools.com/sql/sql_having.asp)

```python
pd.read_sql("""
    select count(name) qty, country    
    from airlines
    where active = 'Y'
    group by country 
    having count(name) = 9 
    """, con=db)
```

#### 10) Which airlines contain the word 'green' somewhere in their name?

Hint: look at the [LIKE](https://www.w3schools.com/sql/sql_like.asp) keyword to see how to search in text columns

```python
pd.read_sql("""
    select name  
    from airlines
    where name like '%green%'

    """, con=db)
```
